```python
import pandas as pd
import numpy as np
```

```python
# input clean data
input_data = pd.read_csv('new_clean_data.csv') 
input_data.head()
```

```python
# create a new column adding other columns up showing numbers of different types toxic contain
input_data['sum'] = input_data.sum(axis=1)
input_data.head()
```

```python
input_data.sort_values(by=['sum'],ascending=False) #sort by numbers of toxic
```

```python
input_data['sum'].value_counts() #value counts summary, 0-6 stand for numbers of toxic 
```

```python
toxic_data = input_data.loc[input_data['sum'] != 0] #output only toxic dataframe
toxic_data.count()#22468
```

```python
toxic_data
```

```python
# n is the numbers of not toxic dataframe will be compiled
n = toxic_data['id'].count()
n 
```

```python
# output not toxic dataframe and shuffle
not_toxic_data = input_data.loc[input_data['sum'] == 0].sample(frac=1)
not_toxic_data.head()
```

```python
not_toxic_data[:n] #extract first n rows from not toxic dataframe
```

```python
# combine into new train dataframe
new_train = toxic_data.append(not_toxic_data[:n]).sort_values(by=['sum'],ascending=True)
new_train.sort_index()
```

```python
new_train.sort_index().to_csv("clean_new_train.csv")
```
