```
import re
import io
import pandas as pd
import string
import numpy as np
```

```
from google.colab import drive
drive.mount('/content/drive')
```

```
train = pd.read_csv('drive/My Drive/jigsaw-toxic-comment-train.csv')
```

```
train.head()
```

```
clean = []
for elem in train.comment_text:
  #remove ip address
  elem = re.sub( r'[0-9]+(?:\.[0-9]+){3}','', elem )
  #remove url
  elem = re.sub(r'http\S+', '', elem)
  #remove tab/newline
  elem = re.sub(r'[\n\t]',' ',elem)
  #remove punctuation
  elem = elem.translate(elem.maketrans("","",string.punctuation))
  #remove hangul
  eng = re.findall(u'[^\uAC00-\uD7A3\u4E00-\u9FA5\u3000-\u303F]',elem)
  for s in elem:
    if s not in eng:
      elem = elem.replace(s,'')
  #remove multiple space
  elem = re.sub(r'[ ]{2,}',' ',elem)
  clean.append(elem.strip())
```

```
clean
```

```
train['comment_text'] = np.array(clean)
```

```
#removing nan and empty strings
count = 0
for elem in train.comment_text:
  if elem == "" or elem ==" ":
    train.drop([count],inplace=True)
  else:
    try: 
      str(elem)
    except:
      train.drop([count],inplace=True)
  count +=1
```
